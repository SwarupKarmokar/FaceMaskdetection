FaceMaskDetector

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
haar_cascade_data = cv2.CascadeClassifier('data.xml')
```

```python
#cv2.rectangle(img,(x,y),(w,h),(b,g,r),border_thickness)
```

```python
capture = cv2.VideoCapture(1)
mask_data = []
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_cascade_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w, y+h), (255,200,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(mask_data))
            if len(mask_data) < 400:
                mask_data.append(face)
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27 or len(mask_data) >= 200:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
np.save('with_mask', mask_data)
```

```python
capture = cv2.VideoCapture(1)
no_mask_data = []
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_cascade_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w, y+h), (255,200,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(no_mask_data))
            if len(no_mask_data) < 400:
                no_mask_data.append(face)
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27 or len(no_mask_data) >= 200:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
np.save('without_mask', no_mask_data)
```
